在距地平線上y=h的高度，考慮雨滴的隨機移動。若向下走一步的機率大於向上走一步的機率，則我們預期此雨滴最後將會落於地平線上。此移動是風向隨機擾動造成雨滴落下的簡單模型（移動機率值為：向上10%，向右15%，向左15%，向下60%）。

# 1.畫出雨滴落下時的高度對時間關係圖


```python
%matplotlib inline
```

```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def raindrop(initial_height):
    
    histogram = list()
   
 #initial_height = 雨滴初始高度，一開始設定100
 #current_height = 雨滴現在高度，一開始跟初始高度一樣，隨時間減少至0 
    
    #initial_height = 100
    current_height = initial_height
    
    state = np.zeros(5)
    
    #state = state + [左,右,雨滴初始高度, 雨滴現在高度,落下所需時間]
    state = state + [0,0,initial_height, current_height,0]
    
    while state[3] > 0:
        p = rnd.random()
        
        #移動機率值為：向上10%，向右15%，向左15%，向下60%
        
                
        #向上移動
        if p >= 0 and p < 0.1:
            state = state + [0,0,0,1,1]
            histogram.append(state)      
        #向下移動
        elif p>=0.1 and p<0.7:
            state = state + [0,0,0,-1,1]
            histogram.append(state)      
        #向左移動
        elif p>=0.7 and p<0.85:
            state = state + [1,0,0,0,1]
            histogram.append(state)      
        #向右移動
        elif p>=0.85 and p<1:
            state = state + [0,1,0,0,1]
            histogram.append(state)    
    return state, histogram


```

```python
#圖為隨機產生一個雨滴從高度=100處落下的軌跡及時間
state, histogram = raindrop(100)
a = list()
b = list()
for i in histogram:
    a.append(i[3])
    b.append(i[4])
fig, ax = plt.subplots()
ax.set_ylabel('height')
ax.set_xlabel('time')
ax.set_title('height-time curve')
plt.plot(b,a, 'bo');
```

# 2.初始高度100，統計雨滴最終狀態的位置分布


執行100次

```python
#圖為隨機產生100個雨滴，雨滴最終落下位置及其數量分布

#final_state = 儲存每個雨滴的最終狀態
final_state = []

for k in range(100):
    f,_ = raindrop(100)
    final_state.append(f)

x = [] #x=儲存每個雨滴最終狀態的位置
for j in range(100):
    x.append(final_state[j][1]-final_state[j][0])

#橫軸(x軸)區間設定在-101~+101之間，單位長度=1
x_ax = np.arange(-101,101,1)
#縱軸(y軸)統計到同一位置的雨滴數
y_ax = np.zeros(len(x_ax))
for i in range(100): #執行100次
    for j in range(202): #長度202
        if x[i] == x_ax[j]:
            y_ax[j] += 1
fig, ax = plt.subplots()
ax.set_ylabel('number')
ax.set_xlabel('position')
ax.set_title('final distribution of raindrops')
ax.hist(x, bins=20);
```
